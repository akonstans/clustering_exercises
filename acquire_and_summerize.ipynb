```python
def null_numbers(df):
    
    cols = ['name', 'rows_missing', 'pct_missing']
    
    null_df = pd.DataFrame(columns= cols)
    
    for col in list(df.columns):
        num_missing = df[col].isna().sum()
        pct_missing = num_missing / len(df)
        
        missing_df = pd.DataFrame([{'name': col, 'rows_missing': num_missing,
                               'pct_missing': pct_missing}])
        
        null_df = pd.concat([null_df, missing_df], axis=0)
        
    null_df.set_index('name', inplace=True)
    
    return null_df
```
